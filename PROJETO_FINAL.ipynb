# Projeto Final — RideSmart

Grupo (individual, dupla, trio e quatro membros).

## Modelagem e Análise de Rotas Urbanas com Grafos

Este notebook é o material-base do projeto final da disciplina.

A ideia é simular um problema inspirado em aplicativos de mobilidade urbana:

> Dado um ponto de origem `A`, um destino `B` e uma distância máxima que o usuário aceita caminhar, escolher o melhor ponto de embarque `P`.

O foco do trabalho tem como meta desenvolver a capacidade de:

- modelar um problema real como grafo;
- definir pesos e critérios de decisão;
- comparar algoritmos de caminhos mínimos;
- analisar criticamente os resultados;
- comunicar os achados de forma clara.

## Problema

Dado:

- ponto de origem `A`;
- ponto de destino `B`;
- distância máxima de caminhada `X`;

o grupo deve escolher um ponto de embarque `P`, tal que `P` esteja a no máximo `X` metros de caminhada da origem `A`.

A rota completa será composta por dois trechos:

```text
A → P  caminhada
P → B  carro
```

O aluno deve comparar:

1. menor rota em distância;
2. rota mais rápida sem trânsito;
3. rota mais rápida com trânsito sintético;
4. caso sem caminhada;
5. ganho obtido ao caminhar até outro ponto de embarque.


# Objetivos de aprendizagem

Ao final do projeto, espera-se que o aluno seja capaz de:

1. representar uma rede urbana real como grafo;
2. usar OSMnx e NetworkX para manipular grafos reais;
3. modelar diferentes funções de custo;
4. comparar algoritmos de caminhos mínimos;
5. interpretar o impacto da escolha dos pesos;
6. discutir o trade-off entre caminhar mais e chegar mais rápido;
7. modelar o trânsito sintético no grafo;
8. avaliar o efeito de trânsito sintético na escolha da rota;
9. apresentar resultados quantitativos e qualitativos.


# Algoritmos que devem ser comparados

Cada grupo deve implementar e comparar, no mínimo:

1. **Dijkstra simples**;
2. **Dijkstra com fila de prioridade / heap**;
3. **A\*** com heurística geográfica;
4. **um algoritmo adicional da literatura para o problema de caminho mais curto**.

O algoritmo adicional deve ser brevemente explicado no Relatório e na apresentação.


# Questões para discussão

No relatório e na apresentação, responda:

1. Como o problema foi modelado como grafo?
2. O que representam os nós e as arestas?
3. Quais pesos foram usados?
4. Como o trânsito sintético alterou as rotas?
5. Caminhar alguns metros melhorou a solução?
6. Em quais casos caminhar atrapalhou?
7. A menor distância foi também a rota mais rápida?
8. O A* expandiu menos nós que o Dijkstra?
9. O Dijkstra com Heap foi mais eficiente que o Dijkstra simples?
10. O algoritmo da literatura trouxe algum ganho?
11. Quais limitações existem na modelagem proposta?
12. Como o modelo poderia ser aproximado de um aplicativo real de mobilidade?
